## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Create a copy of the dataframe to keep the original intact
df_clean = df.copy()

# 1. Handling 'Age' Outliers
# Keep only realistic working ages between 15 and 100
df_clean = df_clean[(df_clean['Age'] >= 15) & (df_clean['Age'] <= 100)]

# 2. Standardizing the 'Gender' Column
# Define lists for grouping the free-text responses
male_str = ["male", "m", "male-ish", "maile", "mal", "male (cis)", "make", "male ", "man","msle", "mail", "malr","cis man", "cis male"]
trans_str = ["trans-female", "something kinda male?", "queer/she/they", "non-binary","nah", "all", "enby", "fluid", "genderqueer", "androgyne", "agender", "male leaning androgynous", "guy (-ish) ^_^", "trans woman", "neuter", "female (trans)", "queer", "ostensibly male, unsure what that really means"]
female_str = ["cis female", "f", "female", "woman",  "femake", "female ","cis-female/femme", "female (cis)", "femail"]

# Map genders to 3 main categories
def clean_gender(gender_str):
    gender_str = str(gender_str).lower().strip()
    if gender_str in male_str:
        return 'Male'
    elif gender_str in female_str:
        return 'Female'
    elif gender_str in trans_str:
        return 'Other'
    else:
        return 'Other' # Catch-all for remaining anomalies

df_clean['Gender'] = df_clean['Gender'].apply(clean_gender)

# 3. Handling Missing Values
# 'work_interfere': Fill nulls with 'Unknown' as missing implies they might not have an illness or don't know
df_clean['work_interfere'].fillna('Unknown', inplace=True)

# 'self_employed': Fill the 18 missing values with the mode (most frequent value, which is 'No')
df_clean['self_employed'].fillna(df_clean['self_employed'].mode()[0], inplace=True)

# 4. Dropping Unnecessary Columns
# 'comments' and 'state' have too many missing values.
# 'Timestamp' is irrelevant for this snapshot analysis.
df_clean.drop(['comments', 'state', 'Timestamp'], axis=1, inplace=True)

# Verify the cleaning
print("Data shape after cleaning:", df_clean.shape)
print("\nRemaining missing values:\n", df_clean.isnull().sum()[df_clean.isnull().sum() > 0])
print("\nCleaned Gender Distribution:\n", df_clean['Gender'].value_counts())

# Overwrite df with df_clean for the rest of the notebook
df = df_clean
````

### What all manipulations have you done and insights you found?

Manipulations Performed:

Outlier Removal (Age): Dropped rows with impossible ages (e.g., negative numbers, ages > 100 or extreme values like 9999). We restricted the dataset to a realistic working-age range of 15 to 100.

Text Standardization (Gender): The original Gender column contained over 40 distinct free-text responses (e.g., "male", "Make", "cis-female/femme", "Enby"). I grouped these into three standardized categories: 'Male', 'Female', and 'Other' using custom mapping lists.

Imputation: >    - work_interfere had 264 missing values. I replaced these with 'Unknown' instead of dropping them, as a non-response is a valid data point (often meaning "Not Applicable").

self_employed had 18 missing values, which were imputed using the mode ('No').

Dimensionality Reduction: Dropped comments (87% missing) and state (highly sparse and US-centric) to reduce noise. Dropped Timestamp as it offers no predictive value for our target.

Insights Found during Wrangling:

Survey data relying on free-text inputs for demographics (like Gender) is highly prone to typos, jokes, and inconsistencies. Standardising this data is absolutely critical before performing any aggregations or visualisations.

After cleaning, we can clearly see that the tech industry sample is heavily skewed toward Males (approx. ~980 respondents), compared to Females (~240) and Others (~20).

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
from sklearn.preprocessing import LabelEncoder

# 1. Create a copy of the dataframe to encode categorical variables into numbers
df_encoded = df.copy()
le = LabelEncoder()

# Loop through all columns and convert 'object' (text) types to numbers
for col in df_encoded.columns:
    if df_encoded[col].dtype == 'object':
        df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))

# 2. Calculate the correlation matrix
corr_matrix = df_encoded.corr()

# 3. Plot the Heatmap
plt.figure(figsize=(18, 14))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm',
            vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation Heatmap of Mental Health Survey Variables', fontsize=16)
plt.show()
````

##### 1. Why did you pick the specific chart?

I chose a Correlation Heatmap because it is the ultimate summary tool for bivariate analysis. It mathematically calculates and visualizes the linear relationships (Pearson correlation) between every single variable in the dataset simultaneously, allowing us to spot hidden relationships that we might have missed in individual bar charts.

##### 2. What is/are the insight(s) found from the chart?

Target Variable Predictors: The strongest correlations with our target variable (treatment) are work_interfere and family_history. This confirms our earlier visual findings that these are the primary drivers of seeking mental health treatment.

Corporate Benefits Cluster: There is a strong positive correlation cluster between benefits, care_options, and wellness_program. If an employer offers one, they usually offer the others, and employees are generally aware of them together.

Stigma Cluster: mental_health_consequence and phys_health_consequence are positively correlated. If an employee feels their company is toxic or penalizes mental health, they usually feel the company will penalize physical health issues as well.
